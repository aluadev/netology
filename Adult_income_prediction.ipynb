```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv( 'adult.csv' )
data.head()
```

```python
data.info()
```

```python
print('Total lines {}'.format( len( data ) ))
```

```python
data = data[ ( data['occupation']=="?") == 0 ]
```

```python
#check the occupation after removing the not required data
data['occupation'].unique()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# choose the features
selectedColumns = data[ [ 'educational-num', 'occupation', 'gender','hours-per-week','income' ] ]

# change to 1 and 0
X = pd.get_dummies( selectedColumns, columns = [ 'occupation', 'gender' ] )

# income is our goal variable, therefore is removed
del X['income']
X.head()
```

```python
# целевая переменная (столбец income) снова является категориальной
# переведем значения столбца в числа, оставив один столбец
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit( data['income'] )
```

```python
le.classes_
```

```python
# записываем в переменную y преобразованный столбец income

y = pd.Series( data = le.transform( data['income'] ) )
y.head()
```

```python
model = LogisticRegression(max_iter=1000)
```

```python
X
```

```python
# train the model with the whole data
model.fit( X, y )
predictions = model.predict_proba( X )
```

```python
model.predict(X)
```

```python
model.score(X, y)
```

```python
pipe_model.fit(X, y)
```

```python
pipe_model.predict(X)
```
